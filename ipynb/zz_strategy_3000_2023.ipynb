```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", None, "display.max_columns", None)

import __init__

from db.pgdf import PgDf
from db.postgresql import PostgreSqlDB

d = PostgreSqlDB()
pg = PgDf()

df = pg.read_sql("""SELECT 
t.result, z.side, 
z.order_id, z.codename, z.dt, zb.proba_all, zb.proba_group, zb.proba_kobetsu
FROM zz_strtg_params as z
JOIN trades as t ON t.order_id = z.order_id
JOIN zz_probas as zb ON zb.order_id = t.order_id
WHERE t.trade_name = 'zzstrat_top3000vol2'
AND abs(t.profit) > 0
AND t.side = 1
AND z.dt >= '2023-01-01 00:00:00'
;""")

df
```

```python
def calculate_dest(row):
    if row['side'] == 1 and row['result'] == 'win':
        return 1
    elif row['side'] == 1 and row['result'] == 'lose':
        return 0
    elif row['side'] == -1 and row['result'] == 'win':
        return 0
    elif row['side'] == -1 and row['result'] == 'lose':
        return 1
    else:
        return 0.5  # or some default value if needed

df['dir'] = df.apply(calculate_dest, axis=1)
df['proba1'] = df['proba_all']*df['proba_group']
df['proba2'] = df['proba_group']*df['proba_kobetsu']
df['proba3'] = df['proba_all']*df['proba_group']*df['proba_kobetsu']
df[['dir', 'proba_all', 'proba_group', 'proba_kobetsu', 'proba1', 'proba2', 'proba3']].corr()
```
