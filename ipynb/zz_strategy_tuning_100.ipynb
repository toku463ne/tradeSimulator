```python
import __init__

import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)

from db.pgdf import PgDf
from db.postgresql import PostgreSqlDB

d = PostgreSqlDB()
pg = PgDf()

df = pg.read_sql("""WITH t AS (
SELECT 
t.profit,
CASE WHEN t.result = 'lose' THEN -1
    WHEN t.result = 'win' THEN 1
    ELSE 0
END AS result,
CASE WHEN (mado > 1.5) THEN 1 ELSE 0 END AS mado,
CASE WHEN (trend_rate < 0.7) THEN 1 ELSE 0 END AS trend,
CASE WHEN ABS(len_std) > 1.5 THEN 1 ELSE 0 END AS long_candle,
prefer_recent_peaks,
CASE WHEN (reversed_cnt > 2) THEN 1 ELSE 0 END AS reversed,
CASE WHEN (ABS(acc) > 0.6) THEN 1 ELSE 0 END AS acc,
ABS(t.takeprofit_price - t.open_price)/ABS(t.stoploss_price - t.open_price) as ts_rate
FROM zz_strtg_params as z
JOIN trades as t ON t.order_id = z.order_id
WHERE t.trade_name = 'zzstrat_top300vol'
AND abs(t.profit) > 0
AND t.open_price > 1000
), g AS (
SELECT COUNT(*) as cnt, result,mado,trend,long_candle,prefer_recent_peaks,reversed, (SUM(t.profit)/20000) as profit
FROM t
WHERE ts_rate > 0.8
GROUP BY result, mado,trend,long_candle,prefer_recent_peaks,reversed
)
SELECT * FROM g ORDER BY 3,4,5,6,7,2
;""")
df
```
