```python
import pandas as pd
import __init__

from db.pgdf import PgDf
from db.postgresql import PostgreSqlDB

d = PostgreSqlDB()
pg = PgDf()

df = None
for (codename,) in d.execSql("SELECT distinct codename from zz_strtg_params;"):
    df_s = pg.read_sql("SELECT * FROM zz_strtg_params WHERE codename = '%s' ORDER BY ep;" % (codename))
    df_t = pg.read_sql("""SELECT 
open_datetime as odt, codename as tcodename, 
CASE WHEN result = 'lose' THEN -1
    WHEN result = 'win' THEN 1
    ELSE 0
END AS res
FROM trades 
WHERE trade_name = 'anal_zigzag' 
AND codename = '%s'
ORDER BY open_epoch, codename;""" % (codename))
    df_tmp = df_t.join(df_s, how='left')
    if df is None:
          df = df_tmp
    else:
        df = pd.concat([df, df_tmp], ignore_index=True)

  
    
```
